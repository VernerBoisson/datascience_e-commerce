```python
import pandas as pd
import yaml
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

with open('../params/configs/data_path.yaml', 'r') as fp:
    data_path = yaml.load(fp)
df_oder_item = pd.read_csv(data_path['olist_order_items_dataset'])
df_product = pd.read_csv(data_path['olist_products_dataset'])
df_orders = pd.read_csv(data_path['olist_orders_dataset'])
translation = pd.read_csv(data_path['product_category_name_translation'])
df = df_orders
df['order_purchase_at'] = pd.to_datetime(df['order_purchase_timestamp'], errors='coerce')
df['order_purchase_at_ym'] = df['order_purchase_at'].dt.to_period('M')
df['order_purchase_at_ymd'] = df['order_purchase_at'].dt.to_period('d')
df.sort_values(by="order_purchase_at_ym", inplace=True)
df.sort_values(by="order_purchase_at_ymd", inplace=True)
```

```python
plt.figure(figsize=(16,10))
sns.countplot(data=df, x='order_purchase_at_ym', palette="Set3")
plt.ylabel("Nombre de commande effectué")
plt.xlabel("Date")
plt.title("Nombre de commande par mois")
plt.xticks(rotation=45)
plt.show()
```

```python
df_tmp = df[df.order_purchase_at_ymd.between('2017-11-01', '2017-12-10')]
fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot()
sns.countplot(data=df_tmp, x='order_purchase_at_ymd', color='grey')
plt.ylabel("Nombre de commande effectué")
plt.xlabel("Date")
ax.patches[23].set_facecolor('darkred')
name_to_color = {
    'Black Friday':   'darkred',
}
patches = [matplotlib.patches.Patch(color=v, label=k) for k,v in name_to_color.items()]
matplotlib.pyplot.legend(handles=patches)
plt.title("Nombre de commande par jour pour le mois de novembre 2017")
plt.xticks(rotation=70)
plt.show()
```

```python
df['order_purchase_at_day'] = df['order_purchase_at'].dt.dayofweek
df.sort_values(by="order_purchase_at_day", inplace=True)
df['order_purchase_at_day'] = df['order_purchase_at_day'].replace(
    {0:"Lun", 1:"Mar", 2:"Mer", 3:"Jeu", 4:"Ven", 5:"Sam", 6:"Dim"})
plt.figure(figsize=(10,10))
sns.countplot(data=df, x='order_purchase_at_day', palette="Set3")
plt.ylabel("Nombre de commande effectué")
plt.xlabel("Jour")
plt.title("Nombre de commande par jour de la semaine")
plt.show()
```

```python

df_customers = pd.read_csv(data_path['olist_customers_dataset'])
df_tmp = pd.merge(df_orders, df_customers, on='customer_id')
df_tmp = pd.merge(df_tmp, df_oder_item, on='order_id')
df_customer_unique = pd.DataFrame(df_tmp.customer_unique_id.value_counts())
df_customer_unique['unique'] = pd.DataFrame(df_tmp.customer_unique_id.value_counts() > 1)
df_customer_unique['str_unique'] = df_customer_unique['unique'].replace(
    {False:"une commande", True:"plusieurs commandes"})
plt.figure(figsize=(10,10))
sns.countplot(data=df_customer_unique, x='str_unique', palette="Set3")
plt.ylabel("Nombre de consommateur")
plt.xlabel('')
plt.title("Nombre de consommateur par rapport au nombre de commande")
plt.show()
```

```python
df_tmp = pd.merge(df_orders, df_customers, on='customer_id')
df_tmp = pd.merge(df_tmp, df_oder_item, on='order_id')
df_tmp = pd.merge(df_tmp, df_product, on='product_id')

df_customer_unique_tmp = df_customer_unique[df_customer_unique.unique]
df_customer_unique_tmp = df_customer_unique_tmp.customer_unique_id[:1000]
list_consumer = df_customer_unique_tmp.index

product_count = df_oder_item['product_id'].value_counts()[:100000]
product_list = product_count.index.tolist()

category_count = df_tmp['product_category_name'].value_counts()[:30]
category_list = category_count.index.tolist()

df_tmp = df_tmp[df_tmp.customer_unique_id.isin(list_consumer)]

df_tmp_1 = df_tmp[df_tmp.product_id.isin(product_list)]

df_tmp_2 = df_tmp[df_tmp.product_category_name.isin(category_list)]


fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot()
sns.scatterplot(data=df_tmp_1, x='customer_unique_id', y='product_id')
plt.xticks(rotation=90)
plt.ylabel('Article')
plt.xlabel('Consommateur')
ax.set_yticks([])
ax.set_xticks([])
plt.title("Répartition des articles par rapport aux consommateurs")
plt.show()


fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot()
sns.scatterplot(data=df_tmp_2, x='customer_unique_id', y='product_category_name')
plt.xticks(rotation=90)
plt.ylabel("Cartégorie d'article")
plt.xlabel('Consommateur')
ax.set_yticks([])
ax.set_xticks([])
plt.title("Répartition des articles par rapport aux consommateurs")
plt.show()
```

```python
df_tmp = pd.merge(df_orders, df_customers, on='customer_id')
df_tmp = pd.merge(df_tmp, df_oder_item, on='order_id')
df_tmp = pd.merge(df_tmp, df_product, on='product_id')
df_agg = df_tmp.groupby(['customer_unique_id']).agg({'customer_unique_id': "count", 'price': "sum"})
df_agg['eco_score'] = df_agg.price/100 * df_agg.customer_unique_id*0.25
df_agg = df_agg.sort_values(['eco_score'], ascending=False)
```

```python
pd.DataFrame(df_agg['eco_score'][:int(df_agg.shape[0]*5/100)])
```

```python
df_tmp = pd.merge(df_orders, df_customers, on='customer_id')

df_reviews = pd.read_csv(data_path['olist_order_reviews_dataset'])

df_tmp = pd.merge(df_tmp, df_reviews, on='order_id')
df_tmp = df_tmp.dropna(subset=['review_score'])
df_agg = df_tmp.groupby(['customer_unique_id']).agg({'customer_unique_id': "count", 
                                                     'review_score': "mean",
                                                     'review_comment_message':"count",
                                                     'review_comment_title':"count",
                                                    })
df_agg['inf_score'] = (df_agg.customer_unique_id * 2  * 
                       df_agg.review_score + 
                       df_agg.review_comment_message * 0.25 + 
                       df_agg.review_comment_message * 0.25) / 4
df_agg = df_agg.sort_values(['inf_score'], ascending=False)
```
